```python
 # Number Guessing Game

while True:
    chances = 3
    
    minNum = input("Enter a Minimum Value : ") 
    maxNum = input("Enter a Maximum Value : ")
    
    if not minNum.isdigit() or not maxNum.isdigit():
        print("Invalid Input Data Type. \nPlease Enter Only Number")
    
    else:
        minNum = int(minNum)
        maxNum = int(maxNum)
    
    if minNum < maxNum:
        secertNum = random.randint(minNum,maxNum)
        
        while chances <= 3:
            print(f"You have {chances} Lives to Guess correct.")
            userGuess = input("Enter Guess Number:")
        
            if not userGuess.isdigit():
                print("Invalid Input Data Type. \nPlease Enter Only Number")
                continue
            else:
                userGuess = int(userGuess)
            
            if userGuess < secertNum:
                print("You Guess Low")
            elif userGuess > secertNum:
                print("You Guess High")
            else:
                print("You Guess Correct Number")
                print("Secert Number", secertNum)
                break 

            chances -= 1

            if chances == 0:
                print(f"You lost all your Chances, Secert Number is {secertNum}")
                break
    else:
        print("Please enter correct values. \nMinimum Number must be less than Maximum Number")
    break
```
